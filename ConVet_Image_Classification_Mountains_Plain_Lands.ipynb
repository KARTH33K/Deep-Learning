```
import os

def files_count(input_path):
  # folder path
  dir_path = str(input_path)
  count = 0
  # Iterate directory
  for path in os.listdir(dir_path):
      # check if current path is a file
      if os.path.isfile(os.path.join(dir_path, path)):
          count += 1
  return count
```

# Convert all images to .jpg

The images scrapped from bing are a mix of different file formats. For building convet models, all images are to be in '.jpg' format. <BR>
The below script is to convert all the scrapped images different than '.jpg' to '.jpg' format.

## Convert mountain images

```
#importing the required package
from PIL import Image
import os



files = os.listdir('/content/images/Mountains/')
for file_name in files:
  filename, file_extension = os.path.splitext(file_name)
  if file_extension == '.jpg':
    continue
  elif file_extension == '.JPG' or file_extension == '.png' or file_extension == '.JPEG' or file_extension == '.jpeg' or file_extension == '.gif':
    img = Image.open('/content/images/Mountains/{}'.format(file_name),'r')
    img = img.convert('RGB')
    img.save('/content/images/Mountains/{}.jpg'.format(filename))
    print('{} is modified to .jpg format'.format(file_name))
    os.remove('/content/images/Mountains/{}'.format(file_name))
  elif file_extension == '.svg':
    drawing = svg2rlg('/content/images/Mountains/{}'.format(file_name))
    renderPM.drawToFile(drawing, "/content/images/Mountains/{}.png".format(filename), fmt="JPG")
    img = Image.open('/content/images/Mountains/{}.JPG'.format(filename))
    img = img.convert('RGB')
    img.save('/content/images/Mountains/{}.jpg'.format(filename))
    print('{} is modified to .jpg'.format(file_name))
    os.remove('/content/images/Mountains/{}'.format(file_name))
```

# Data Augmentation 

```
train_dir = '/content/images/train'
```

```
test_dir = '/content/images/test'
```

Define custom batch size:

```
Custom_Batch_Size = 10
```

```
from keras.preprocessing.image import ImageDataGenerator

# All images will be rescaled by 1./255
train_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
        # This is the target directory
        train_dir,
        # All images will be resized to 150x150
        target_size=(150, 150),
        batch_size=Custom_Batch_Size,
        # Since we use binary_crossentropy loss, we need binary labels
        class_mode='binary')

test_generator = test_datagen.flow_from_directory(
        test_dir,
        target_size=(150, 150),
        batch_size=Custom_Batch_Size,
        class_mode='binary')
```

## Build Model

```
from keras import layers
from keras import models

model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu',
                        input_shape=(150, 150, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Flatten())
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
```

# Model Compile

```
from keras import optimizers

model.compile(loss='binary_crossentropy',
              optimizer='Adam',
              metrics=['acc'])
```

# Model Fit

```
import time

start = time.time()
print("hello")

history = model.fit(
      train_generator,
      steps_per_epoch= (files_count('/content/images/train/mountains')+files_count('/content/images/train/non_mountains'))//Custom_Batch_Size, # len(trainX) // BS == 500/10  == 50
      epochs=50,
      validation_data=test_generator,
      validation_steps= ((files_count('/content/images/test/mountains')+files_count('/content/images/test/non_mountains'))//Custom_Batch_Size))  # ideally...len(valX) // BS == 400/10 == 40

end = time.time()
print("\n Start time: ",start)
print("End time: ",end)
print("Exxecution time: ",(end - start))
```

# Plots - Training/Validation acuracy, loss

```
import matplotlib.pyplot as plt

acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(len(acc))

plt.plot(epochs, acc, 'bo', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.legend()

plt.figure()

plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.legend()

plt.show()
```

# Model Performance Insights & Observations



1.   Data preparation for a convoluted neural network is the most challenging and time-consuming activity compared to building and tuning the model. Roughly 80% of the time is spent on coding for fetching images, converting non-jpg images to jpg format, and creating folder structure for the model to access the images.<BR><BR>
2.   Accuracy reached around 90% without data augmentation. But with data augmentation, the accuracy was 100%.<BR><BR>
3.   With a batch size of 10 for CNN models, accuracy reached 100% within 25-30 epochs. When the batch size is reduced, the model takes lesser epochs to reach that accuracy. However, validation loss is bigger for smaller batch sizes than for bigger batch sizes. Also, this might could also be because of the small dataset used for the model. Therefore, run the model with a large dataset to observe if CNN obtains the maximum accuracy in a short number of epochs.<BR><BR>
4.   Given the time and machine specifications limitations, I copied the same images used for training into the test folder in the same order. However, you can test by extracting random images from the web and copying them into the test folder.<BR><BR>
5.   The F1 Score indicates that images are the only requirement for modeling. No additional inputs or data are required to be modeled for CNN. The more perfect images we feed to the model, the better the prediction by the model because the model entirely relies on the images.
